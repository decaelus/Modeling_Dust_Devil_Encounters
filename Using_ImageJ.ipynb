```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from scipy.optimize import curve_fit
```

```python
# Let's create a function to model and create data
def func(x, baseline, a, x0, sigma):
    return baseline - a*np.exp(-(x-x0)**2/(2*sigma**2))
```

```python
data = np.genfromtxt('/Users/brian/Downloads/Values2.csv', delimiter=",", names=True)
scale = 11.05 # m/px
```

```python
ind = ~np.isnan(data['X0']) & ~np.isnan(data['Y0'])
x = data['X0'][ind]
y = data['Y0'][ind]

plt.plot(x, y)
popt, pcov = curve_fit(func, x, y, p0=[140., 30., 7.5, 2.5])
print(popt)
print("optical depth: ", np.log(popt[0]/(popt[0] - popt[1])))
print("diameter: ", scale*popt[3])
plt.plot(x, func(x, *popt))
```

```python
# Let's create a function to model and create data
def positive_gaussian(x, baseline, a, x0, sigma):
    return baseline + a*np.exp(-(x-x0)**2/(2*sigma**2))
```

```python
data = np.genfromtxt('/Users/brian/Downloads/SecondDD.csv', delimiter=",", names=True)

ind = ~np.isnan(data['X']) & ~np.isnan(data['Y'])
x = data['X'][ind]
y = data['Y'][ind]

plt.plot(x, y)
popt, pcov = curve_fit(positive_gaussian, x, y, p0=[np.mean(y), 30., np.mean(x), 2.5])
print(popt)
print("optical depth: ", np.log(popt[0]/(popt[0] - popt[1])))
print("diameter: ", scale*popt[3])
plt.plot(x, positive_gaussian(x, *popt))
```
